```python
import pandas as pd
```

```python
QC_input = pd.read_csv('QC_input.csv')
QC_output = pd.read_csv('QC_output.csv')
```

```python
def QC_module(mturk_res):
    resume_col_names = [ ".".join(col.split(".")[1:]) for col in QC_input.columns if col.split(".")[0] == 'GS'][1:]
    gold_standard_cols = ["GS." + col for col in resume_col_names]
    gold_standard_truth_cols = ["GStruth." + col for col in resume_col_names]
    for i in range(len(gold_standard_cols)):
        QC_input["GS pass?"] = QC_input[gold_standard_cols[i]] == QC_input[gold_standard_truth_cols[i]]
    dropped = gold_standard_cols + gold_standard_truth_cols
    return QC_input[QC_input["GS pass?"] == True].drop(columns=dropped)
```
